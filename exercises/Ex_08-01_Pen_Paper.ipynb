### Aufgabe 8.1

Zeigen Sie, dass nach Gl. (8.15) die Korrelation durch eine Faltung ausgedrückt
werden kann.

Die Notation in Gl. (8.15) ist etwas irreführend. Für einen vorgegebenen time-lag $\tau$ können wir die Faltung der kontinuierlichen Funktionen $x(-t)$ und $y(t)$ folgendermaßen schreiben:

\begin{equation}
(x(-t) \ast y(t))(\tau) = \int_{-\infty}^{\infty} x(-t')y(\tau-t')dt'\notag
\end{equation}

Mit der Substitution $u = -t'$ folgt 

\begin{equation}
dt' = - du\notag
\end{equation}

sowie neue Integrationsgrenzen

\begin{equation}
t' = -\infty \rightarrow u = \infty \notag
\end{equation}

\begin{equation}
t' = \infty \rightarrow u = -\infty \notag.
\end{equation}

Damit folgt 

\begin{equation}
(x(-t) \ast y(t))(\tau) = \int_{-\infty}^{\infty} x(-t')y(\tau-t')dt' = -\int_{\infty}^{-\infty} x(u)y(\tau+u)du\notag
\end{equation}

oder

\begin{equation}
(x(-t) \ast y(t))(\tau) = -\int_{\infty}^{-\infty} x(u)y(\tau+u)du = \int_{-\infty}^{\infty} x(u)y(\tau+u)du = \rho_{x,y}(\tau),\notag
\end{equation}

was zu zeigen war.

\rho_{x,x}(\tau) = \int_{-\infty}^{\infty} [a\delta(t')+b\delta(t'-t_0)] [a\delta(t'+\tau)+b\delta(t'+\tau-t_0)] dt' \notag

### Aufgabe 8.3

Zeigen Sie, dass die AKF einer Sinusfunktion $\sin(\omega_0 t)$, die auf einem endlichen
Intervall $T = 2k\pi/\omega_0$, $k \in \mathbb{N}$ gegeben ist, näherungsweise eine Cosinusfunktion ist (Gl. 8.27). Hinweise: Nutzen Sie ein Additionstheorem. In welchem Integrationsintervall ist das Korrelationsintegral ungleich Null?

Die Aufgabenstellung ist etwas mehrdeutig. Eigentlich soll die AKF der Sinusfunktion berechnet werden. Der Verweis auf Gl. 8.27 gilt aber für die KKF einer Sinusfunktion 

$x(t)=\begin{cases}sin(\omega_0 t)&:t\in[0,T]\\0&:\text{ sonst}\end{cases}$

mit der Kreisfrequenz $\omega_0$ und einer phasenverschobenen Sinusfunktion: 

$y(t)=\begin{cases}sin(\omega_0 t + \phi)&:t\in[0,T]\\0&:\text{ sonst}\end{cases}$

mit der Phasenverschiebung $\phi$.

Da die AKF ein Spezialfall der KKF für $\phi = 0$ ist, berechne ich im folgenden die KKF. Dabei müssen wir zwei Fälle unterscheiden:

##### Positive Lags $T\ge\tau\ge0$

In diesem Fall verschieben wir die Funktion y(t) relativ zur Funktion x(t) nach "links" auf der Zeitachse, d.h. für ein vorgegebenes $T\ge\tau\ge0$ ergeben sich nur auf dem Integrationsintervall $[0,T-\tau]$ des Korrelationsintegrals Werte ungleich Null:

\begin{equation}
\rho_{x,y}(\tau) = \frac{1}{T} \int_0^{T-\tau} sin(\omega_0 t')\; sin(\omega_0 (t'+\tau) + \phi)\; dt' \notag
\end{equation}

Als erstes vereinfachen wir das Argument der Sinusfunktionen, indem wir 

\begin{equation}
u := \omega_0 t' \notag
\end{equation}

substituieren. Damit ändert sich das Integrationsintervall von 

\begin{equation}
t' \in [0,T-\tau] \notag
\end{equation}

zu 

\begin{equation}
u \in [0,\omega_0(T-\tau)] \notag
\end{equation}

$dt'$ wird folgendermaßen durch $du$ substituiert:

\begin{equation}
dt' = \frac{1}{\omega_0} du \notag
\end{equation}

Damit vereinfacht sich das Korrelationsintegral zu 

\begin{equation}
\rho_{x,y}(\tau) = \frac{1}{\omega_0 T}\int_0^{\omega_0(T-\tau)} sin(u)\; sin(u + \omega_0 \tau + \phi)\; du \notag
\end{equation}

Im nächsten Schritt wäre es praktisch, wenn wir das Produkt der Sinusfunktionen durch eine Summe von Winkelfunktionen ersetzen könnten. Eine kurze Suche bei Wikipedia ergibt im Abschnitt [Formelsammlung_Trigonometrie#Produkte_der_Winkelfunktionen](https://de.wikipedia.org/wiki/Formelsammlung_Trigonometrie#Produkte_der_Winkelfunktionen) folgende Beziehung:

\begin{equation}
sin(x)sin(y) = \frac{1}{2}(cos(x-y)-cos(x+y))\notag
\end{equation}

Damit folgt für unser Korrelationsintegral:

\begin{equation}
\rho_{x,y}(\tau) = \frac{1}{2\omega_0 T}\int_0^{\omega_0(T-\tau)} cos(-(\omega_0 \tau + \phi)) - cos(2u + \omega_0 \tau + \phi) \; du \notag
\end{equation}

Nutzen wir die Achsensymmetrie der Kosinusfunktion $cos(-x)=cos(x)$ folgt

\begin{equation}
\rho_{x,y}(\tau) = \frac{1}{2\omega_0 T}\int_0^{\omega_0(T-\tau)} cos(\omega_0 \tau + \phi) - cos(2u + \omega_0 \tau + \phi) \; du \notag
\end{equation}

Die Stammfunktion des Integraden ergibt sich einfach zu:

\begin{equation}
\rho_{x,y}(\tau) = \frac{1}{2\omega_0 T}\biggl[u\; cos(\omega_0 \tau + \phi) - \frac{sin(2u + \omega_0 \tau + \phi)}{2}\biggr]_0^{\omega_0(T-\tau)} \notag
\end{equation}

Auswertung der Stammfunktion an den Intervallgrenzen ergibt:

\begin{equation}
\rho_{x,y}(\tau) = \frac{1}{2\omega_0 T}\biggl[\omega_0(T-\tau)\; cos(\omega_0 \tau + \phi) - \frac{sin(2\omega_0(T-\tau) + \omega_0 \tau + \phi)}{2}+\frac{sin(\omega_0 \tau + \phi)}{2}\biggr] \notag
\end{equation}

oder nach einigen Vereinfachungen:

\begin{equation}
\rho_{x,y}(\tau) = \frac{(T-\tau)}{2T}\; cos(\omega_0 \tau + \phi) - \frac{sin(2\omega_0 T - \omega_0 \tau + \phi)}{4\omega_0 T}+\frac{sin(\omega_0 \tau + \phi)}{4\omega_0 T} \notag
\end{equation}

Im zweiten Term auf der RHS können wir ausnutzen, daß laut Definition die Funktion periodisch auf dem Intervall T sein soll. Damit können wir $2\omega_0 T = 0$ setzen:

\begin{equation}
\rho_{x,y}(\tau) = \frac{(T-\tau)}{2T}\; cos(\omega_0 \tau + \phi) - \frac{sin(- \omega_0 \tau + \phi)}{4\omega_0 T}+\frac{sin(\omega_0 \tau + \phi)}{4\omega_0 T} \notag
\end{equation}

Mit der Punktsymmetrie der Sinusfunktion $sin(-x)=-sin(x)$ folgt 

\begin{equation}
\rho_{x,y}(\tau) = \frac{(T-\tau)}{2T}\; cos(\omega_0 \tau + \phi) + \frac{sin(\omega_0 \tau - \phi)}{4\omega_0 T}+\frac{sin(\omega_0 \tau + \phi)}{4\omega_0 T} \notag
\end{equation}

Im Fall von $\phi = \frac{\pi}{2} + n \pi$ mit $n \in \mathbb{N}_0$ heben sich die beiden Sinusfunktionen durch destruktive Interferenz auf, so daß nur der Kosinusterm übrig bleibt. 

Für große $\omega_0$ und $T$ ist der Beitrag der beiden Sinusfunktionen auf der RHS selbst im Fall einer beliebigen Phasenverschiebung $\phi$ sehr klein. Damit dominiert der Kosinusterm und man kann annehmen, daß für $T\ge\tau\ge0$ näherungsweise gilt

\begin{equation}
\rho_{x,y}(\tau) \approx \frac{(T-\tau)}{2T}\; cos(\omega_0 \tau + \phi)\notag
\end{equation}

##### Negative Lags $-T\le\tau\le0$

Im Fall von negativen Lags verschieben wir die Funktion y(t) relativ zur Funktion x(t) nach "rechts" auf der Zeitachse, d.h. für ein vorgegebenes $-T\le\tau\le0$ ergeben sich nur auf dem Integrationsintervall $[|\tau|,T]$ des Korrelationsintegrals Werte ungleich Null:

\begin{equation}
\rho_{x,y}(\tau) = \frac{1}{T} \int_{|\tau|}^{T} sin(\omega_0 t')\; sin(\omega_0 (t'+\tau) + \phi)\; dt' \notag
\end{equation}

Im Vergleich zu dem Fall $\tau \ge 0$ ändert sich am Integranden und damit der Stammfunktion nichts, wir müssen diese nur an den neuen Intregationsgrenzen auswerten. D.h. es gilt:

\begin{equation}
\rho_{x,y}(\tau) = \frac{1}{2\omega_0 T}\biggl[u\; cos(\omega_0 \tau + \phi) - \frac{sin(2u + \omega_0 \tau + \phi)}{2}\biggr]_{\omega_0 |\tau|}^{\omega_0 T} \notag
\end{equation}

und damit 

\begin{equation}
\rho_{x,y}(\tau) = \frac{1}{2\omega_0 T}\biggl[\omega_0 (T-|\tau|)\; cos(\omega_0 \tau + \phi) - \frac{sin(2\omega_0 T + \omega_0 \tau + \phi)}{2} + \frac{sin(2\omega_0 |\tau| + \omega_0 \tau + \phi)}{2}\biggr] \notag
\end{equation}

oder nach einigen Vereinfachungen

\begin{equation}
\rho_{x,y}(\tau) = \biggl[\frac{(T-|\tau|)}{2T}\; cos(\omega_0 \tau + \phi) - \frac{sin(2\omega_0 T + \omega_0 \tau + \phi)}{4 \omega_0 T} + \frac{sin(2\omega_0 |\tau| + \omega_0 \tau + \phi)}{4 \omega_0 T}\biggr] \notag
\end{equation}

Nutzen wir erneut die Periodizität der Funktionen $x(t)$ und $y(t)$ aus, ergibt sich mit $2\omega_0 T = 0$

\begin{equation}
\rho_{x,y}(\tau) = \biggl[\frac{(T-|\tau|)}{2T}\; cos(\omega_0 \tau + \phi) - \frac{sin(\omega_0 \tau + \phi)}{4 \omega_0 T} + \frac{sin(2\omega_0 |\tau| + \omega_0 \tau + \phi)}{4 \omega_0 T}\biggr] \notag
\end{equation}

Für große $\omega_0$ und $T$ kann die Näherung

\begin{equation}
\rho_{x,y}(\tau) \approx \frac{(T-|\tau|)}{2T}\; cos(\omega_0 \tau + \phi) \notag
\end{equation}

verwendet werden.  

Somit gilt:

\begin{equation}
\rho_{x,y}(\tau) \approx \frac{(T-\tau)}{2T}\; cos(\omega_0 \tau + \phi)\; \text{für}\; T\ge\tau\ge0 \notag
\end{equation}

\begin{equation}
\rho_{x,y}(\tau) \approx \frac{(T-|\tau|)}{2T}\; cos(\omega_0 \tau + \phi)\; \text{für}\; -T\le\tau\le0 \notag
\end{equation}

oder allgemein 

\begin{equation}
\rho_{x,y}(\tau)=\frac{1}{T}\int\limits_{0}^{T}x(t^{\prime})y(t^{\prime}+\tau)\mathrm{dt^{\prime}}\approx
\begin{cases}
\frac{T-|\tau|}{2T}\cos(\omega_{0}\tau+\phi)&:\tau\in[-T,T]\\
0&:\text{ sonst}
\end{cases}\notag
\end{equation}

was zu zeigen war.

In der folgenden Funktion `KKF_xy` will ich den Unterschied zwischen der exakten Lösung von $\rho_{x,y}$, der Näherung und der numerischen Lösung miteinander vergleichen. 

```python
# Importiere Python Bibliotheken 
# ------------------------------
%matplotlib inline
from ipywidgets import interactive
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal
from gsv.gsv_func import *        # Bibliothek "Geophysikalische Signalverarbeitung
```

```python
def KKF_xy(phi,f0,T,sw):
    
    # Define sine function x on time interval T
    # -----------------------------------------
    time, x = create_sine_int(f0,T,0.)   # Create sine function on [0,T] with phi = 0.
    time, y = create_sine_int(f0,T,phi)  # Create sine function on [0,T] with phi ne 0.
    
    # KKF(x,y)
    KKF_xy = signal.correlate(y, x, mode='same')

    # Estimate sample interval dt and mulitply KKF_xy with dt to get correct amplitudes
    dt = time[1] - time[0]
    KKF_xy = KKF_xy * dt / T

    # Compute time-lags
    nt = len(KKF_xy)                    # number of samples
    lags = np.arange(-nt/2,nt/2,1) * dt # time-lags
    
    # Analytical KKF
    # --------------
    w0 = 2. * np.pi * f0    # circular frequency
    KKF_an = np.zeros(nt)
    
    if(sw==1): # approximation
        
        KKF_an = 0.5 * (T-np.abs(lags)) * np.cos(w0*lags+phi) / T
        
    if(sw==2): # exact
        
        # loop over lags
        for i in range(nt):
            
            if(lags[i]>=0.): # use exact solution for lags >= 0
                KKF_an1 = (T-lags[i]) * np.cos(w0*lags[i]+phi) / (2*T)
                KKF_an2 = - np.sin(2.*w0*(T-lags[i]/2)+phi) / (4.*T*w0) 
                KKF_an3 = np.sin(w0*lags[i]+phi) / (4.*T*w0)
                KKF_an[i] = KKF_an1 + KKF_an2 + KKF_an3
            
            if(lags[i]<0.): # use exact solution for lags < 0
                KKF_an1 = (T-np.abs(lags[i])) * np.cos(w0*lags[i]+phi) / (2*T)
                KKF_an2 = - np.sin(2.*w0*T + w0*lags[i]+phi) / (4.*T*w0) 
                KKF_an3 = np.sin(2.*w0*np.abs(lags[i])+w0*lags[i]+phi) / (4.*T*w0)
                KKF_an[i] = KKF_an1 + KKF_an2 + KKF_an3
        
    if(sw==3): # difference exakt - approximation
        
        # loop over lags
        for i in range(nt):
            
            # KKF difference exact - approx for ...
            if(lags[i]>=0.): # ... exact solution (lags >= 0)
                KKF_an2 = - np.sin(2.*w0*(T-lags[i]/2)+phi) / (4.*T*w0) 
                KKF_an3 = np.sin(w0*lags[i]+phi) / (4.*T*w0)
                KKF_an[i] = KKF_an2 + KKF_an3
            
            if(lags[i]<0.): # ... exact solution (lags < 0)
                KKF_an2 = - np.sin(w0*lags[i]+phi) / (4.*T*w0) 
                KKF_an3 = np.sin(2.*w0*np.abs(lags[i])+w0*lags[i]+phi) / (4.*T*w0)
                KKF_an[i] = KKF_an2 + KKF_an3

    # set AKF for lags > abs(T) to zero
    nth = (int)(len(lags)/2)
    dt = lags[-1] - lags[-2]
    nT = (int)(T/dt)
    
    KKF_an[0:nth-nT] = 0.0
    KKF_an[nth+nT:-1] = 0.0
    KKF_an[-1] = 0.0
    
    # Plot time series x(t), y(t) and their KKF_xy
    plt.figure(figsize=(20,10))
    
    plt.subplot(211)
    
    plt.plot(time, x, 'b-',lw=3,label='x(t)')
    plt.plot(time, y, 'r-',lw=3,label='y(t)')
    plt.title('x(t), y(t)')
    plt.xlabel('time [s]')
    plt.ylabel('Sine function x(t), y(t)')
    plt.legend()
    plt.grid()
    
    plt.subplot(212)
    
    # plot KKF
    plt.plot(lags, KKF_xy, 'b-',lw=3,label='numeric')
    if(sw==1):
        plt.plot(lags, KKF_an, 'r--',lw=3,label='analytical (approx)')
    if(sw==2):
        plt.plot(lags, KKF_an, 'r--',lw=3,label='analytical (exact)')
    if(sw==3):
        plt.plot(lags, KKF_an, 'r--',lw=3,label='analytical (exact-approx)')
    
    plt.title('KKF(x,y)')
    plt.xlabel(r'time-lags $\tau$ [s]')
    plt.ylabel('KKF(x,y)')
    plt.legend()
    plt.grid()
```

Eingabeparameter der Funktion `KKF_xy` sind die **Phasenverschiebung $\phi\in[0,2\pi]$**, welche in $\frac{\pi}{4}$ Schritten variiert werden kann. Ebenso können die **Frequenz $f_0\in[1\;Hz,16\;Hz]$** der Sinusfunktion und **Fensterlänge $T\in[0\;s,2\;s]$** variiert werden. Schließlich kann mit dem Parameter **sw** zwischen der Näherungslösung Gl. (8.27) (**sw=1**), der exakten Lösung (**sw=2**) und der Differenz zwischen exakter und Näherungslösung (**sw=3**) gewechselt werden.

```python
interactive_plot = interactive(KKF_xy, phi=(0, 2.*np.pi, np.pi/4.), f0=(1., 16., 1.), T=(.5, 2.,.5), sw=(1,3,1))
output = interactive_plot.children[-1]
output.layout.height = '350px'
interactive_plot
```

Durch Variation der Parameter kann man folgende Beobachtungen machen:

- Für $\phi=\frac{\pi}{2}+n \pi$ mit $n\in\mathbb{N}_0$ erkennt man, daß die Abweichungen zwischen exakter und Näherungslösung Null werden.

- Die Abweichungen zwischen exakter und Näherungslösung werden groß, wenn man $\phi = 0$, $f_0 = 1\;Hz$ und $T = 0.5\; s$ wählt. 

- Umgekehrt werden die Abweichungen sehr klein, wenn man $\phi = 0$, $f_0 = 16\;Hz$ und $T = 2\; s$ wählt.
