# Polarisationsanalyse an einer 3-Komponentenstation

Aus der Analyse von 3-Komponentendaten lassen sich Einfallswinkel unterschiedlicher seismischer Phasen relativ zu einer Station, sowie deren Polarisationseigenschaften berechnen. Diese Informationen können genutzt werden, um Aussagen über die seismische Anisotropie im Untergrund in unmittelbarer Umgebung der Station abzuleiten. In Kombination mit Laufzeitunterschieden verschiedener Wellenphasen, z.B. P- und S-Wellen, lassen sich seismische Quellen grob lokalisieren.

## Theoretische Grundlagen

Die Partikelbewegung an einer 3-Komponentenstation $\vec{u}(t)$ lautet im Frequenzbereich:

\begin{equation}
\mathfrak{F}\{\vec{u}(t)\}=\vec{U}(\omega)=
\begin{pmatrix}
U_{z}(\omega)\\
U_{N}(\omega)\\
U_{W}(\omega)
\end{pmatrix}
\notag
\end{equation}

Für das Spektrum $\vec{U}$ wird als Vektor $\vec{p}$, welcher die Polarisation der Welle beschreibt, multipliziert mit einem Skalar $V$, welcher die Anregung beschreibt, angenommen. Dabei kann die Polarisation $\vec{p}$ als ein durch die Matrix $A$ rotierter Vektor $\vec{p_0}$ aufgefasst werden, wobei $\vec{p_0}$ ein bekanntes Modell für die Polarisation des Wellentyps angibt:

\begin{equation}
\vec{U}(\omega)=\vec{p}(\omega)\cdot V(\omega)=A(\omega)\cdot\vec{p_{0}}(\omega)\cdot V(\omega)
\notag
\end{equation}

Für eine P-Welle im isotropen Vollraum wäre:

$\vec{p_{0}}=\begin{pmatrix}1\\0\\0\end{pmatrix}$. 

Abhängig von der Ausbreitungsrichtung, aber auch beeinflusst durch die freie Oberfläche, Heterogenitäten und Anisotropie im Untergrund, treten Abweichungen vom isotropen Polarisationsvektor auf. 

Für eine Rayleigh-Welle im isotropen Halbraum mit Elliptizität $\varepsilon(\omega)$ gilt 

$\vec{p_{0}}=\begin{pmatrix}1\\0\\i\varepsilon(\omega)\end{pmatrix}$

**Bestimmung von $\vec{p}(\omega)$:**

1. Bestimmung der Korrelationsmatrix $C(\omega)$:

\begin{align}
c_{ij}(\omega)&=\rho_{u_i,u_j}(\omega)=U_i^{\ast}(\omega)U_j(\omega)=p_i^{\ast}(\omega)p_j^{\top}(\omega)|V(\omega)|^{2} \notag\\
\Leftrightarrow C(\omega)&=\vec{p}^{\ast}(\omega)\vec{p}^{\top}(\omega)|V(\omega)|^{2}
\notag
\end{align}

2. Normierung von $C(\omega)$:

\begin{equation}
\tilde{C}(\omega)=\frac{C(\omega)}{\mathrm{tr}(C(\omega))}=\tilde{\vec{p}}^{\ast}(\omega)\tilde{\vec{p}}^{\top}(\omega)
\notag
\end{equation}

3. [Singulärwertzerlegung](https://en.wikipedia.org/wiki/Singular_value_decomposition) von $\tilde{C}(\omega)$:

\begin{equation}
\mathrm{SVD}(\tilde{C}(\omega))=E^{\ast}(\omega)\lambda(\omega)E^{\top}(\omega)
\notag
\end{equation}

Dabei bezeichnet, im Fall einer diagonalizierbaren Matrix $\tilde{C}(\omega)$, $\lambda$ die Matrix der Eigenwerte und $E=(\vec{e_1},\vec{e_2},\vec{e_3})$ die Matrix der Eigenvektoren. Aus den Eigenvektoren läßt sich die Polarisation $\tilde{\vec{p}}$ ablesen und entsprechend dem Modell des Wellentyps interpretieren:

$\tilde{\vec{p}}$  für die **P-Welle** zeigt in Ausbreitungsrichtung

$\tilde{\vec{p}}$  für die **S-Welle** zeigt senkrecht zur Ausbreitungsrichtung

$\tilde{\vec{p}}$  für die **Rayleighwelle** ist komplex und die horizontalen Komponenten zeigen die Ausbreitungsrichtung an
