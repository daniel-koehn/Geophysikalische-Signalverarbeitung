```python
# Importiere Python Bibliotheken 
# ------------------------------
%matplotlib inline
from ipywidgets import interactive
import matplotlib.pyplot as plt
import numpy as np
```

Als erstes müssen wir den komprimierten Felddatensatz im Ordner `data` entpacken. 

Um die Felddaten einzulesen, muß nur die Funktion `create shotgather` leicht modifiziert werden. Die Daten werden aus einer ASCII Datei eingelesen, welche das Shotgather bereits in Form einer Matrix enthält. Wichtige Akquisitionsparamter wie der Abstand der Empfänger und das zeitliche Sampleintervall werden in dem Notebook definiert.

```python
def create_shotgather():
    
    # Load Cachtice SH-data (profile 2)
    TT = np.loadtxt('data/Cachtice_prof2_y.shot1', delimiter='\t', skiprows=0, unpack='True')
    nx, nt = np.shape(TT) # estimate nt and nx
    
    # Define acquisition parameters
    dx = 0.5                 # spatial geophone sampling [m]
    xmax = nx * dx           # maximum offset [m]
    
    dt = 1.25e-4             # time sampling [s]    
    tmax = nt * dt           # maximum recording time [s]
    
    # Define offsets and t
    x = np.arange(0, xmax, dx) # offsets [m]
    t = np.arange(0, tmax, dt) # time [s]
    
    return x,t,TT
```

```python
x,t,TT = create_shotgather()

# Initialize Plots
plt.figure(figsize=(10,7))
    
# Plot shot gather
clip=5e0
plt.imshow(TT.T, cmap=plt.cm.seismic, interpolation='none', vmin=-clip, vmax=clip, 
                extent=[np.min(x),np.max(x),np.max(t),np.min(t)],aspect=1e1)
plt.xlabel('Offset (m)')
plt.ylabel('Time (s)')
title = r'Shot Gather'
plt.title(title)
```

Auffällige Einsätze in dem Shotgather sind die direkte Welle, eine refraktierte und eine Oberflächenwelle vom Love Typ. Darüberhinaus sind schwache Reflexionen und Diffraktionen in den Daten erkennbar. 

### NMO Korrektur 

Wie man sieht kann man mit dem Konvolutionsmodell einen einfachen synthetischen Datensatz erzeugen. Im nächsten Schritt soll der **Normal Move Out (NMO)** der direkten Welle für verschiedene Testgeschwindigkeiten $vstack$ über die Formel 

\begin{equation}
\Delta t_{NMO} = - \frac{x}{vstack}\notag
\end{equation}

korrigiert und die korrigierten seismischen Spuren gemäß 

\begin{equation}
\mbox{stack}(t) = \frac{1}{M} \sum_{i=1}^M x_i(t+\Delta t_{NMO}). \notag
\end{equation}

aufgestapelt werden.

```python
def NMO_corr(vstack,stackcorr):
    
    # create data
    x,t,TT = create_shotgather()
    
    # Define parameters
    nx = len(x)
    nt = len(t)
    dt = t[1]-t[0]
    
    if(stackcorr==True):
        
        # NMO correction by dtNMO = - offset / vstack 
        # --------------------------------------------------------
        TTc = np.zeros((nx,nt))  # initialize corrected shot gather
        tstack = (x / vstack)    # traveltime [s]
        nc = (tstack/dt)         # traveltime [sample#]
        nc = nc.astype(np.int32) # convert floats to integers
    
        for i in range(0,nx):
            for j in range(0,nt):
            
                if((j+nc[i]) < nt):
                    TTc[i,j] = TT[i,j+nc[i]]
    
        
        TT = TTc
        
    # Initialize Plots
    plt.figure(figsize=(10,5))
    
    plt.subplot(121)
    
    # Plot shot gather
    clip=5e0
    plt.imshow(TT.T, cmap=plt.cm.seismic, interpolation='none', vmin=-clip, vmax=clip, 
                   extent=[np.min(x),np.max(x),np.max(t),np.min(t)],aspect=2e1)
    plt.xlabel('Offset (m)')
    plt.ylabel('Time (s)')
    title = r'Shot Gather'
    plt.title(title)
    
    plt.subplot(122)
    
    # Stack data along offsets
    stack = np.sum(TT,0) / nx
    
    # Plot stack
    plt.plot(np.flip(stack), np.flip(t), 'b')
    plt.xlabel('Stack Amplitude ()')
    plt.title('Stack')
    plt.xlim(-1.,1.)
    plt.ylim(np.min(t),np.max(t))
    plt.gca().invert_yaxis()
```

```python
interactive_plot = interactive(NMO_corr, vstack=(10., 400., 10.), stackcorr=False )
output = interactive_plot.children[-1]
output.layout.height = '350px'
interactive_plot
```

Für die direkte Welle erhalten wir eine brauchbare Korrektur des NMO bei $vstack_d=100\; m/s$, für die Love-Welle bei ca. $vstack_l = 130\; m/s$ und für die refraktierte Welle bei $vstack_r = 220\; m/s$.

### Slant Stack 

Basierend auf der Funktion `NMO_corr` können wir eine Funktion programmieren, welche den **Slant Stack** eines Shot Gathers berechnet. Dazu wenden wir einfach die NMO Korrektur für verschiedene Stapelgeschwindigkeiten $vstack$ auf das Shot Gather an und speichern die gestapelten Stacks für jede Geschwindigkeit. Plotten wir die Zeiten über die Stacks erhalten wir den Slant Stack ...

```python
def slant_stack():
    
    # create data
    x,t,TT = create_shotgather()
    
    # Define parameters
    nx = len(x)
    nt = len(t)
    dt = t[1]-t[0]
    
    # Compute Slant Stack
    # -------------------
    
    # define array for stacking velocities
    vstack = np.arange(10.,500.,10.)
    nv = len(vstack) # number of tested velocities
    
    # initialize slant stack with zeros
    sstack = np.zeros((nv,nt))
    
    # loop over velocities
    for k in range(0,nv) :
        
        # NMO correction by dtNMO = - offset / vstack 
        TTc = np.zeros((nx,nt))    # initialize corrected shot gather
        tstack = (x / vstack[k])   # traveltime [s]
        nc = (tstack/dt)           # traveltime [sample#]
        nc = nc.astype(np.int32)   # convert floats to integers
    
        for i in range(0,nx):
            for j in range(0,nt):
            
                if((j+nc[i]) < nt):
                    TTc[i,j] = TT[i,j+nc[i]]
    
        # Stack NMO corrected data
        stack = np.sum(TTc,0) / nx
            
    
        # update slant stack
        sstack[k,:] = stack[:]
    
    
    # Initialize Plots
    plt.figure(figsize=(8,8))
    
    # Plot slant stack
    plt.imshow(sstack.T, cmap=plt.cm.seismic, interpolation='none', 
                   extent=[np.min(vstack),np.max(vstack),np.max(t),np.min(t)],aspect=5e2)
    plt.xlabel('vstack (m/s)')
    plt.ylabel(r'$\tau$ (s)')
    title = r'Slant Stack'
    plt.title(title)
    
    return sstack
```

```python
sstack = slant_stack()
```

Basierend auf unserer "interaktiven" Geschwindigkeitsbestimmung mit dem Vergleich zwischen Shotgather und Stack im vorherigen Abschnitt, können wir die Maxima im Slant Stack entsprechend interpretieren. Direkte Welle $vstack_d=100\; m/s$, Love-Welle $vstack_l = 130\; m/s$ und die refraktierte Welle $vstack_r = 220\; m/s$.

### Phasengeschwindigkeits-Frequenzspektren

Berechnen wir die Fouriertransformation der Zeitachse erhalten wir das **Phasengeschwindigkeits-Frequenzspektrum**. Dies läßt sich einfach in unseren Slant-Stack Code implementieren ...

```python
def pf_spec():
    
    # create data
    x,t,TT = create_shotgather()
    
    # Define parameters
    nx = len(x)
    nt = len(t)
    dt = t[1]-t[0]
    
    # Compute Slant Stack
    # -------------------
    
    # define array for stacking velocities
    vstack = np.arange(10.,300.,10.)
    nv = len(vstack) # number of tested velocities
    
    # initialize slant stack with zeros
    sstack = np.zeros((nv,nt))
    
    # loop over velocities
    for k in range(0,nv) :
        
        # NMO correction by dtNMO = - offset / vstack 
        TTc = np.zeros((nx,nt))    # initialize corrected shot gather
        tstack = (x / vstack[k])   # traveltime [s]
        nc = (tstack/dt)           # traveltime [sample#]
        nc = nc.astype(np.int32)   # convert floats to integers
    
        for i in range(0,nx):
            for j in range(0,nt):
            
                if((j+nc[i]) < nt):
                    TTc[i,j] = TT[i,j+nc[i]]
    
        # Stack NMO corrected data
        stack = np.sum(TTc,0) / nx
            
        # update slant stack
        sstack[k,:] = stack[:]
    
    # Compute pf-spectrum
    # -------------------
    
    # Fourier transform slant stack along the time axis
    pf = np.fft.fft(sstack)
    
    # Normalize by length of the time series
    pf = pf / (nt*dt)
    
    # estimate frequencies    
    freq = np.fft.fftfreq(nt, d=dt)
    
    # Initialize Plots
    plt.figure(figsize=(8,8))
    
    # Plot pf spectrum
    levels = np.linspace(0., np.max(np.abs(pf)), num=1000)
    plt.contourf(freq,vstack,np.abs(pf),levels=levels,cmap='magma')
    
    plt.ylabel('Phase velocity vstack (m/s)')
    plt.xlabel(r'Frequency (Hz)')
    title = r'Phasevelocity-Frequency Spectrum'
    plt.title(title)
    plt.xlim(0.,60.)
```

```python
pf_spec()
```

Das dominante Signale im Phasengeschwindigkeit-Frequenzspektrum ist die Love Welle, deren Geschwindigkeit im Frequenzbereich zwischen 10 Hz und 30 Hz von 250 m/s bis 100 m/s variiert. Da die tieferfrequenten Anteile der Love-Welle tiefer in den Untergrund eindringen als die hochfrequenten, können wir daraus folgern, daß die S-Wellengeschwindigkeit im Untergrund entsprechend zunimmt.
